```python
file_path = "abfss://raw@retaildatasg.dfs.core.windows.net/product_details.csv"
product_df = spark.read.csv(file_path, header=True, inferSchema=True)
display(product_df)
```

```python
bronze_path = "abfss://bronze@retaildatasg.dfs.core.windows.net/product_details/"

product_df.write \
    .mode("overwrite") \
    .parquet(bronze_path)
```

```python
display(
    dbutils.fs.ls(
        "abfss://bronze@retaildatasg.dfs.core.windows.net/product_details/"
    )
)
```
